```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
Anime = ['Go-Toubun No Hanayome', 'Rascal Does Not Dream Of Bunny Girl Senpai', 'Your Lie In April', 'Plastic Memories']
Sufrimiento =   [60,60,100,80]

plt.figure(figsize=(8, 4))
plt.bar(Anime, Sufrimiento, color = ['blue', 'black', 'yellow', 'purple'])

plt.title('Lista de animes por sufrimiento', fontsize=12, fontweight='bold')
plt.xlabel('Animes', fontsize=12)
plt.ylabel('porcentaje de Sufrimiento (%)', fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.show()
```
