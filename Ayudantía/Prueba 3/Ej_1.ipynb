```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
Lenguajes = ['Python', 'C++', 'Java', 'JavaScript', 'C#']
Preferencias = [45, 20, 15, 12, 8]

print(Lenguajes, 
      Preferencias)
```

```python
plt.figure(figsize=(4, 4))
plt.bar(Lenguajes, Preferencias, color=['indigo', 'orange', 'red', 'gold', 'teal'])

plt.title('Preferencia de Lenguajes de Programación 2026', fontsize=12, fontweight='bold')
plt.xlabel('Lenguajes de Programación', fontsize=12)
plt.ylabel('Porcentaje de Preferencia (%)', fontsize=12)

plt.show()
```
