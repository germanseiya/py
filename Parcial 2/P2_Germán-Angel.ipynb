```python
Nombre = ['P Groudon', 'M blaziken' ,'G Darmanitan Z-M' ,'Ho-oh' ,'M Charizard X' , 'Blacephalon','Emboar', 'Blaziken', 'Reshiram', 'M Camerupt']
NivelAtaque = [180, 160, 160, 130, 130, 127,123, 120, 120, 120]

plt.figure(figsize=(16, 8))
plt.bar(Nombre, NivelAtaque, color=['indigo', 'orange', 'red', 'gold', 'teal'])

plt.title('Top 10 pokemons tipo fuego por su nivel de ataque', fontsize=12, fontweight='bold')
plt.xlabel('Nombre', fontsize=12)
plt.ylabel('Nivel de Ataque', fontsize=12)

plt.show()
```
